```python
import pandas as pd
from ydata_profiling import ProfileReport
```

# Dataset 1: Financial Phrase Bank – Domain-Specific Sentiment

```python
# =========================================================
# Dataset 1: Financial Phrase Bank – Domain-Specific Sentiment
# =========================================================
print("="*60)
print("DATASET 1: Financial Phrase Bank")
print("="*60)

url_financial = "https://github.com/santluan/tcc-mba-usp/blob/main/dados/financial_sentences.csv?raw=true"
df_financial = pd.read_csv(url_financial)
```

```python
# Manual EDA (what students usually do)
print(" MANUAL EDA - Financial Dataset")
print(f"Shape: {df_financial.shape}")
print(f"Number of variables: {len(df_financial.columns)}")
print(f"Missing cells: {df_financial.isnull().sum().sum()}")
print(f"Duplicate rows: {df_financial.duplicated().sum()} "
      f"({df_financial.duplicated().mean()*100:.1f}%)")
print("\nInfo:")
print(df_financial.info())
print("\nDescribe:")
print(df_financial.describe(include='all'))
```

```python
# Automated EDA with YData Profiling
print("\n Generating YData Profiling Report for Financial Dataset...")
profile_financial = ProfileReport(
    df_financial,
    title="Financial Phrase Bank Report",
    explorative=True,
    minimal=False
)
profile_financial.to_notebook_iframe()

print("\n Key insights from YData Profiling (Financial):")
print("- Check 'Overview' tab for: Number of variables, Missing cells %, Duplicates")
print("- Look at 'Alerts' for any data quality issues")
print("- Sentiment distribution and text statistics are also available\n")
```

```python
# Display in notebook + Save to HTML file
profile_financial.to_notebook_iframe()
profile_financial.to_file("financial_phrase_bank_eda_report.html")

print(" Financial report saved as: financial_phrase_bank_eda_report.html")
```

# Dataset 2: Twitter Sentiment Analysis

```python
# =========================================================
# Dataset 2: Twitter Sentiment Analysis
# =========================================================
print("="*60)
print("DATASET 2: Twitter Sentiment Analysis")
print("="*60)

url_twitter = "https://raw.githubusercontent.com/dD2405/Twitter_Sentiment_Analysis/master/train.csv"
df_twitter = pd.read_csv(url_twitter)
```

```python
# Manual EDA
print(" MANUAL EDA - Twitter Dataset")
print(f"Shape: {df_twitter.shape}")
print(f"Number of variables: {len(df_twitter.columns)}")
print(f"Missing cells: {df_twitter.isnull().sum().sum()}")
print(f"Duplicate rows: {df_twitter.duplicated().sum()} "
      f"({df_twitter.duplicated().mean()*100:.1f}%)")
print("\nInfo:")
print(df_twitter.info())
print("\nDescribe:")
print(df_twitter.describe(include='all'))
```

```python
# Automated EDA with YData Profiling
print("\n Generating YData Profiling Report for Twitter Dataset...")
profile_twitter = ProfileReport(
    df_twitter,
    title="Twitter Sentiment Report",
    explorative=True,
    minimal=False
)
profile_twitter.to_notebook_iframe()

print("\n Key insights from YData Profiling (Twitter):")
print("- Check 'Overview' tab for: Number of variables, Missing cells %, Duplicates")
print("- 'tweet' column shows rich text analysis (length, words, etc.)")
print("- 'label' column shows class distribution (very important for sentiment models)")
print("- Alerts tab highlights any hidden data quality problems\n")
```

```python
# Display in notebook + Save to HTML file
profile_twitter.to_notebook_iframe()
profile_twitter.to_file("twitter_sentiment_eda_report.html")

print(" Twitter report saved as: twitter_sentiment_eda_report.html")
```
