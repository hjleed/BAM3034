## Part 1: Sentiment Analysis with VADER

VADER is a rule-based sentiment analysis tool that relies on a lexicon of sentiment-related words, their intensity, and some rules for modifying the scores based on context (e.g., capitalization, punctuation).

### Step 1: Install Required Libraries

To use VADER for sentiment analysis, we first need to install the vaderSentiment library. Additionally, we will use nltk for downloading the VADER lexicon.

### Step 2: Sentiment Analysis Using VADER

```python
# Import necessary libraries
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Sample sentences for sentiment analysis
texts = [
    "I love this product! It's amazing.",
    "This is the worst thing I have ever bought.",
    "It's okay, but not great.",
    "Absolutely fantastic! Highly recommend it.",
    "I'm not happy with the customer service."
]

# Analyzing sentiment
for text in texts:
    score = analyzer.polarity_scores(text)
    print(f"Text: {text}")
    print(f"Sentiment Scores: {score}")
    print(f"Sentiment: {'Positive' if score['compound'] >= 0.05 else 'Negative' if score['compound'] <= -0.05 else 'Neutral'}")
    print("\n")
```

**Explanation:**

* **VADER Sentiment Score**: VADER returns a dictionary with four values:
  * **positive**: Proportion of text that is positive.
  * **neutral**: Proportion of text that is neutral.
  * **negative**: Proportion of text that is negative.
  * **compound**: A normalized score that sums up the sentiment. It ranges from -1 (most negative) to +1 (most positive). If it's above 0.05, it's considered positive, below -0.05 is negative, and between -0.05 and 0.05 is neutral.

## Part 2: Sentiment Analysis Using Huggingface Transformers

Huggingface offers a wide range of pre-trained models, including sentiment analysis models like BERT or DistilBERT, which are fine-tuned on sentiment analysis tasks. These models provide a deep understanding of sentiment based on context and word relationships.

### Step 2: Sentiment Analysis Using Huggingface's Pre-trained Model

We will use the DistilBERT model fine-tuned for sentiment analysis, which is smaller, faster, and efficient while still performing well.

```python
from transformers import pipeline

# Load the sentiment analysis pipeline from Huggingface
sentiment_pipeline = pipeline("sentiment-analysis")

# Sample sentences for sentiment analysis
texts = [
    "I love this product! It's amazing.",
    "This is the worst thing I have ever bought.",
    "It's okay, but not great.",
    "Absolutely fantastic! Highly recommend it.",
    "I'm not happy with the customer service."
]

# Analyzing sentiment using Huggingface model
for text in texts:
    result = sentiment_pipeline(text)[0]
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}, Score: {result['score']:.4f}")
    print("\n")
```

**Explanation:**

* The Huggingface Sentiment Analysis Pipeline uses a pre-trained model that outputs a label (LABEL_0 for negative and LABEL_1 for positive sentiment) and a confidence score.
* The model is fine-tuned to understand sentiment with much greater nuance than simple lexicon-based models like VADER.

```python
# ================================================
# SARCASM TEST - VADER vs Hugging Face
# ================================================

# List of sarcastic sentences 
sarcastic_examples = [
    "I love how this app crashes every five minutes.",
    "Wonderful, another flight delay that ruins everything.",
    "Amazing work, you managed to make it worse.",
    "Perfect, now the whole project is failing.",
    "This is just great, everything is going wrong."
]

print(" BATCH SARCASM TEST - VADER vs Hugging Face")
print("="*75)

results_summary = []

for i, text in enumerate(sarcastic_examples, 1):
    print(f"\n Example {i}: \"{text}\"")

    # -------- VADER --------
    vader_score = analyzer.polarity_scores(text)
    compound = vader_score['compound']
    if compound >= 0.05:
        vader_sentiment = "Positive"
    elif compound <= -0.05:
        vader_sentiment = "Negative"
    else:
        vader_sentiment = "Neutral"

    # -------- Hugging Face --------
    hf_result = sentiment_pipeline(text)[0]
    hf_label = hf_result['label']
    hf_conf = hf_result['score']

    # -------- Print Results --------
    print(f"   VADER → {vader_sentiment:8} (compound = {compound:.3f})")
    print(f"   BERT  → {hf_label:8} (confidence = {hf_conf:.1%})")

    results_summary.append((vader_sentiment, hf_label))

print("\n" + "="*75)

# -------- Summary Comparison --------
vader_positive = sum(1 for v, _ in results_summary if v == "Positive")
hf_negative = sum(1 for _, h in results_summary if h == "NEGATIVE")

print("\n SUMMARY:")
print(f"• VADER labeled {vader_positive}/{len(sarcastic_examples)} sarcastic sentences as **Positive**")
print(f"• BERT labeled {hf_negative}/{len(sarcastic_examples)} sarcastic sentences as **NEGATIVE**")
print("\n Key Insight:")
print("   VADER is fooled by positive words like 'love', 'wonderful', 'amazing', 'perfect', 'great'.")
print("   Hugging Face (DistilBERT) understands the sarcastic intent and correctly flags almost all as NEGATIVE.")
print("   → This is a classic example of why transformer models outperform rule-based tools on real-world text!")
```
