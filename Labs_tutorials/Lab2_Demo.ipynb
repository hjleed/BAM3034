```python
import re

slang_dict = {
    "N-words": "(**)",
    "dam-words": "(**)",
    "f-words": "(**)"
}

def replace_slang(text):
    for slang, replacement in slang_dict.items():
        text = re.sub(rf"\b{slang}\b", replacement, text, flags=re.IGNORECASE)
    return text

def parse_data(data):
    # Split the raw text into movie sections
    sections = data.split("NewMovieDrPQRd")
    parsed = []

    for section in sections:
        section = section.strip()
        if not section:
            continue

        # Extract movie ID (assuming format: MovieID:12345)
        movie_id_match = re.search(r"MovieID\s*:\s*(\d+)", section)
        movie_id = movie_id_match.group(1) if movie_id_match else None

        # Extract comments (assuming they follow "Comments:" until next delimiter)
        comments_match = re.search(r"Comments\s*:(.*)", section, flags=re.DOTALL)
        comments = comments_match.group(1).strip() if comments_match else ""

        # Clean comments
        comments = replace_slang(comments)

        parsed.append({
            "movie_id": movie_id,
            "comments": comments
        })

    return parsed
```
